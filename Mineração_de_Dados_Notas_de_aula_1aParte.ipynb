#### RegressÃ£o:
A regressÃ£o Ã© um tipo de algoritmo de aprendizado supervisionado usado para prever um valor numÃ©rico contÃ­nuo. Funciona analisando a relaÃ§Ã£o entre uma variÃ¡vel dependente e uma ou mais variÃ¡veis ââindependentes. O objetivo da regressÃ£o Ã© encontrar a melhor linha ou curva de ajuste que possa prever a variÃ¡vel dependente com base nos valores das variÃ¡veis ââindependentes. A regressÃ£o pode ser usada para vÃ¡rios fins, como prever preÃ§os de aÃ§Ãµes, preÃ§os de imÃ³veis ou nÃºmeros de vendas.

```python
# Generate some random data
x = np.array([2, 4, 6, 8, 10, 11, 6, 12, 13, 14, 11])
y = np.array([6, 7, 11, 13, 17, 18, 15, 21, 23, 25, 26])

# Fit a linear regression line
coefficients = np.polyfit(x, y, 1)
m = coefficients[0]
b = coefficients[1]

# Create a scatter plot of the data points
plt.scatter(x, y)

# Add the regression line to the plot
plt.plot(x, m*x + b, color='red')

# Add axis labels and a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

# Display the plot
plt.show()
```

#### AssociaÃ§Ã£o:
AssociaÃ§Ã£o Ã© um tipo de algoritmo de aprendizado nÃ£o supervisionado usado para identificar padrÃµes ou relacionamentos entre variÃ¡veis ââem um conjunto de dados. Ele funciona encontrando conjuntos de itens frequentes ou conjuntos de itens que ocorrem simultaneamente no conjunto de dados com mais frequÃªncia do que o esperado por acaso. O objetivo da associaÃ§Ã£o Ã© descobrir relaÃ§Ãµes interessantes entre variÃ¡veis, como quais itens sÃ£o frequentemente comprados juntos em uma anÃ¡lise de cesta de compras.



```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(42)
data = []
products = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
for i in range(40):
    transaction = np.random.choice(products, size=10, replace=True)
    data.append(transaction)

print("Baskets:")
for x in data:
  print(set(x))

# Convert the data to a one-hot encoded matrix
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Generate frequent itemsets with a minimum support of 0.2
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)

# Generate association rules with a minimum confidence of 0.7
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.83)

print("\nMined rules")
print(rules)

# Create a scatter plot of the association rules
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Association Rules')


rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(list(x))).astype("unicode")
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(list(x))).astype("unicode")

for i in range(len(rules)):
    text = rules['antecedents'][i] + " => " + rules['consequents'][i]
    if rules['confidence'][i] < 0.9:
      plt.annotate(text, (rules['support'][i], rules['confidence'][i]+i*0.005))
    else:
      plt.annotate(text, (rules['support'][i], rules['confidence'][i]-i*0.005))

# Display the plot
plt.show()
```

#### Tabelas

A maneira mais simples e rudimentar de representar a saÃ­da do aprendizado de mÃ¡quina Ã© tornÃ¡-la igual Ã  entrada - uma tabela. Caso o processo seja usado para previsÃ£o numÃ©rica chamamos de tabela de regressÃ£o. De forma menos trivial, a criaÃ§Ã£o de uma tabela de decisÃ£o ou regressÃ£o pode envolver a seleÃ§Ã£o de alguns dos atributos. Se a temperatura for irrelevante para a decisÃ£o de jogar ou nÃ£o golfe, por exemplo, uma tabela menor e condensada com esse atributo ausente seria um guia melhor. O problema Ã©, claro, decidir quais atributos deixar de fora sem afetar a decisÃ£o final.

```python
import itertools
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

# Create a dataset with the famous weather/play golf problem
data = pd.DataFrame({
    'outlook': ['sunny', 'sunny', 'overcast', 'rain', 'rain', 'rain', 'overcast', 'sunny', 'sunny', 'rain', 'sunny', 'overcast', 'overcast', 'rain'],
    'temperature': ['hot', 'hot', 'hot', 'mild', 'cool', 'cool', 'cool', 'mild', 'cool', 'mild', 'mild', 'mild', 'hot', 'mild'],
    'humidity': ['high', 'high', 'high', 'high', 'normal', 'normal', 'normal', 'high', 'normal', 'normal', 'normal', 'high', 'normal', 'high'],
    'wind': ['weak', 'strong', 'weak', 'weak', 'weak', 'strong', 'strong', 'weak', 'weak', 'weak', 'strong', 'strong', 'weak', 'strong'],
    'play_golf': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
})

print(data)

# Encode the nominal attributes as integers
label_encoder = LabelEncoder()
for col in data.columns:
    data[col] = label_encoder.fit_transform(data[col])

# Create a list of all 36 possible combinations of nominal attributes
outlook_values = ['sunny', 'overcast', 'rain']
temperature_values = ['hot', 'mild', 'cold']
humidity_values = ['high', 'normal']
wind_values = ['weak', 'strong']
complete_table = list(itertools.product(* 
                      [outlook_values, 
                       temperature_values, 
                       humidity_values, 
                       wind_values]))
cdf = pd.DataFrame(columns=['outlook', 'temperature', 'humidity', 'wind'])
for x in complete_table: cdf.loc[len(cdf)] = x
for col in cdf.columns:
    cdf[col] = label_encoder.fit_transform(cdf[col])

X = data[['outlook', 'temperature', 'humidity', 'wind']]
y = data['play_golf']
clf = DecisionTreeClassifier()
clf.fit(X, y)

res = clf.predict(cdf)
print()
print('Model results:')
for index in range(len(complete_table)):
  print(complete_table[index], " -> ", 'yes' if res[index] == 1 else 'no')

```

#### Regras

Geralmente, as condiÃ§Ãµes sÃ£o combinadas logicamente de maneira conjuntiva (E) e todos os testes devem ser bem-sucedidos para que a regra seja acionada. No entanto, em algumas formulaÃ§Ãµes de regras, as condiÃ§Ãµes sÃ£o expressÃµes lÃ³gicas gerais, em vez de simples conjunÃ§Ãµes.

Frequentemente pensamos nas regras individuais como sendo efetivamente logicamente disjuntivas (OU): se qualquer uma se aplica, a classe (ou distribuiÃ§Ã£o de probabilidade) fornecida em sua conclusÃ£o Ã© aplicada Ã  instÃ¢ncia. #### Ãrvores de decisÃ£o

Uma abordagem de âdividir e conquistarâ para o problema de aprender a partir de um conjunto de instÃ¢ncias independentes leva naturalmente a um estilo de representaÃ§Ã£o chamado decisÃ£o Ã¡rvore. 

NÃ³s em uma Ã¡rvore de decisÃ£o envolvem o teste de um atributo especÃ­fico.  Normalmente, o teste compara um valor de atributo com uma constante. Os nÃ³s folha fornecem uma classificaÃ§Ã£o que se aplica a todas as instÃ¢ncias que atingem a folha, ou um conjunto de classificaÃ§Ãµes, ou uma distribuiÃ§Ã£o de probabilidade sobre todas as classes possÃ­veis. 

Para classificar uma instÃ¢ncia desconhecida, ela Ã© roteada para baixo da Ã¡rvore de acordo com os valores dos atributos testados em nÃ³s sucessivos,e quando uma folha Ã© alcanÃ§ada a instÃ¢ncia Ã© classificada de acordo com a classe atribuÃ­da Ã  folha.

Se o atributo testado em um nÃ³ for nominal, o nÃºmero de filhos geralmente Ã© o nÃºmero de valores possÃ­veis do atributo. Nesse caso, como existe um ramo para cada valor possÃ­vel, o mesmo atributo nÃ£o serÃ¡ ser testado novamente mais abaixo na Ã¡rvore. Ãs vezes, os valores dos atributos sÃ£o divididos
em dois subconjuntos, e a Ã¡rvore se ramifica apenas de duas maneiras, dependendo de qual subconjunto o valor estÃ¡ na Ã¡rvore; nesse caso, o atributo pode ser testado mais de uma vez em um caminho.

Se o atributo for numÃ©rico, o teste em um nÃ³ geralmente determina se seu
valor Ã© maior ou menor que uma constante predeterminada, dando uma divisÃ£o de duas vias. Alternativamente, uma divisÃ£o de trÃªs vias pode ser usada, caso em que existem vÃ¡rios diferentes possibilidades. Uma alternativa para um atributo de valor inteiro seria uma divisÃ£o tripla em menor que, igual a e maior que. Uma alternativa para um atributo de valor real, para o qual a igualdade nÃ£o Ã© tÃ£o significativa, seria testar contra um intervalo em vez de uma Ãºnica constante, novamente dando uma divisÃ£o de trÃªs vias: abaixo, dentro e acima. Um atributo numÃ©rico Ã© frequentemente testado vÃ¡rias vezes em qualquer caminho da Ã¡rvore desde a raiz atÃ© a folha, cada teste envolvendo uma constante diferente. 



```python
from sklearn.tree import plot_tree
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.inspection import DecisionBoundaryDisplay

iris = load_iris()

plt.rcParams["figure.figsize"] = (10,7.5)
plt.figure()
clf = DecisionTreeClassifier().fit(iris.data, iris.target)
plot_tree(clf, filled=True)
plt.title("Decision tree trained on all the iris features")
plt.show()
```

#### Modelos lineares

Os estatÃ­sticos usam a palavra regressÃ£o para o processo de prever uma quantidade numÃ©rica e "modelo linear de regressÃ£oâ Ã© outro termo para esse tipo de modelo. Infelizmente isso nÃ£o se relaciona com o uso comum da palavra âregressÃ£oâ, que significa retornar a um estado anterior.

Os modelos lineares tambÃ©m podem ser aplicados a problemas de classificaÃ§Ã£o binÃ¡ria. A partir de um conjunto de dados, uma certa quantia Ã© mantida para teste - isso Ã© chamado de **procedimento de validaÃ§Ã£o** - e o restante usado para

#### Deixar um de fora

A validaÃ§Ã£o cruzada *leave-one-out* Ã© simplesmente uma validaÃ§Ã£o cruzada $n$ vezes, onde $n$ Ã© o nÃºmero de instÃ¢ncias no conjunto de dados. ### *Bootstrap*

O *bootstrap*, Ã© baseado no procedimento estatÃ­stico de amostragem com reposiÃ§Ã£o. Anteriormente, sempre que uma amostra era retirada do conjunto de dados para formar um conjunto de treinamento ou teste, ela era retirada do conjunto. A ideia do bootstrap Ã© amostrar o conjunto de dados com substituiÃ§Ã£o para formar um conjunto de treinamento. Descreveremos uma variante particular, misteriosamente (mas por uma razÃ£o que logo se tornarÃ¡ aparente) chamado de **bootstrap $0.632$**. Como alguns elementos neste segundo conjunto de dados serÃ£o (quase certamente) repetidos, deve haver algumas instÃ¢ncias no conjunto de dados original que nÃ£o foram escolhidas: vamos usÃ¡-las como instÃ¢ncias de teste.
Qual Ã© a chance de uma determinada instÃ¢ncia nÃ£o ser escolhida para o treinamento?

Ele tem uma probabilidade de $\frac{1}{n}$ de ser escolhida a cada vez e, portanto, uma probabilidade de $\left( 1 - \frac{1}{n} \right)$ de nÃ£o ser escolhida. Como este processo Ã© repetido $n$ vezes de forma independente, a chance de nÃ£o ser escolhida ao longo do processo Ã© (para $n$ grande o suficiente):

$p_{fora} = \left( 1 - \frac{1}{n} \right)^n â e^{-1} \approx 0.36787944117...$

O complementar deste valor Ã© a probabilidade de ser escolhida, ou seja:

$p_{dentro} = 1 - p_{fora} = 0.63212055882...$

Assim, para um conjunto de dados razoavelmente grande, o conjunto de teste conterÃ¡ cerca de 36,8% das instÃ¢ncias e o conjunto de treinamento conterÃ¡ cerca de 63,2% delas. Algumas instÃ¢ncias serÃ£o repetidas no conjunto de treinamento, trazendo atÃ© um tamanho total de n, o mesmo que no conjunto de dados original.

Como apenas 63,2% das instÃ¢ncias sÃ£o usadas na elaboraÃ§Ã£o do modelo, este pode tornar-se um modelo fraco em comparaÃ§Ã£o a uma validaÃ§Ã£o cruzada de 10 partes que utiliza cerca de 90% das instÃ¢ncias para treinamento. Como forma de compensar esta deficiÃªncia, o erro de resubstituiÃ§Ã£o Ã© trazido no cÃ¡lculo de uma estimativa de erro mais justa. Portanto, para uma validaÃ§Ã£o via *bootstrap* a taxa de erro final Ã© reequilibrada como:

$e = 0.632 \cdot e_{teste} + 0.368 \cdot e_{treinamento}$

EntÃ£o, todo o procedimento bootstrap Ã© repetido vÃ¡rias vezes, com diferentes
amostras de substituiÃ§Ã£o para o conjunto de treinamento e a mÃ©dia dos resultados.

O *bootstrap* pode ser considerado como o mÃ©todo mais adequado para a utilizaÃ§Ã£o em pequenas quantidades de dados ou dados cujas classes sejam altamente desequilibradas em suas representatividade.

### Vizinhos mais prÃ³ximos (KNN)

Embora o aprendizado baseado em instÃ¢ncias seja simples e eficaz, geralmente Ã© lento. A maneira Ã³bvia de descobrir qual membro do conjunto de treinamento estÃ¡ mais prÃ³ximo de uma instÃ¢ncia de teste desconhecida Ã© calcular a distÃ¢ncia de cada membro do conjunto de treinamento e selecionar o menor. Este procedimento Ã© linear no nÃºmero de instÃ¢ncias de treinamento: ou seja, o tempo necessÃ¡rio para fazer uma Ãºnica previsÃ£o Ã© proporcional ao nÃºmero de instÃ¢ncias de treinamento. 

O processamento de um conjunto de teste inteiro leva um tempo proporcional ao produto do nÃºmero de instÃ¢ncias nos conjuntos de treinamento e teste. Os vizinhos mais prÃ³ximos podem ser encontrados de forma mais eficiente representando o conjunto de treinamento como uma Ã¡rvore, embora nÃ£o seja muito Ã³bvio como. Uma estrutura adequada Ã© uma kDtree.

Esta Ã© uma Ã¡rvore binÃ¡ria que divide o espaÃ§o de entrada com um hiperplano e depois divide cada partiÃ§Ã£o novamente, recursivamente. Todas as divisÃµes sÃ£o feitas paralelamente a um dos eixos, seja na vertical ou na horizontal, no caso bidimensional. A estrutura de dados Ã© chamada de kD-tree porque armazena um conjunto de pontos no espaÃ§o k-dimensional, sendo k o nÃºmero de atributos.

O aprendizado baseado na instÃ¢ncia do vizinho mais prÃ³ximo Ã© simples e geralmente funciona muito bem. A priori cada atributo tem exatamente a mesma influÃªncia sobre a decisÃ£o, assim como no mÃ©todo Naive Bayes. Outro problema Ã© que o banco de dados pode ser facilmente corrompido por exemplares ruidosos (*outliers*). Uma soluÃ§Ã£o Ã© adotar a estratÃ©gia de k-vizinho mais prÃ³ximo, onde algum nÃºmero fixo, pequeno, k de vizinhos mais prÃ³ximos - digamos cinco - sÃ£o localizados e usados juntos para determinar a classe da instÃ¢ncia de teste por meio de uma votaÃ§Ã£o de maioria simples. (Observe que anteriormente usamos k para
denotam o nÃºmero de atributos; este Ã© um uso diferente e independente.) 

O mÃ©todo do vizinho mais prÃ³ximo surgiu hÃ¡ muitas dÃ©cadas, e os estatÃ­sticos
analisaram esquemas de k-vizinhos mais prÃ³ximos no inÃ­cio dos anos 1950. Se o nÃºmero instÃ¢ncias de treinamento Ã© grande, faz sentido usar mais de um vizinho mais prÃ³ximo, mas claramente isso Ã© perigoso se houver poucos casos. Pode-se mostrar que quando k e o nÃºmero n de instÃ¢ncias se tornam infinitos de tal forma que $\frac{k}{n} \rightarrow 0$, a probabilidade de erro se aproxima do mÃ­nimo teÃ³rico para o conjunto de dados. 

O mÃ©todo do vizinho mais prÃ³ximo foi adotado como um esquema de classificaÃ§Ã£o no inÃ­cio dÃ©cada de 1960 e tem sido amplamente utilizado no campo de reconhecimento de padrÃµes por quase metade de um sÃ©culo.

A classificaÃ§Ã£o do vizinho mais prÃ³ximo era notoriamente lenta atÃ© que as Ã¡rvores kD comeÃ§aram a ser aplicadas no inÃ­cio dos anos 1990, embora a prÃ³pria estrutura de dados tenha sido desenvolvida muito mais cedo. Na prÃ¡tica, essas Ã¡rvores tornam-se ineficientes quando a dimensÃ£o do espaÃ§o aumenta e sÃ³ valem a pena quando o nÃºmero de atributos Ã© relativamente pequeno. Ãrvores de bola foram desenvolvidas muito mais recentemente e sÃ£o uma instÃ¢ncia
de uma estrutura mais geral chamada Ã¡rvore mÃ©trica.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier

# Generate some random data with 2 classes
X, y = make_blobs(n_samples=200, centers=2, random_state=0)

# Fit a KNN classifier with k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)

# Create a mesh grid to plot the decision boundary
xmin, xmax = X[:, 0].min() - 1, X[:, 0].max() + 1
ymin, ymax = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(xmin, xmax, 0.1),
                     np.arange(ymin, ymax, 0.1))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Classify a new example and plot its neighbors
new_example = [3, 2.5]
distances, indices = knn.kneighbors([new_example])
fig, ax = plt.subplots()
ax.contourf(xx, yy, Z, alpha=0.4)
ax.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
ax.scatter(new_example[0], new_example[1], marker='x', s=100, color='red')
for i in indices[0]:
    ax.plot([new_example[0], X[i, 0]], [new_example[1], X[i, 1]], 'k--', alpha=0.3)
ax.set_title('K-Nearest Neighbors Classification with k=5')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
plt.show()
```

### Naive Bayes

O mÃ©todo 1R usa um Ãºnico atributo como base para suas decisÃµes e escolhe aquele que funciona melhor. Outra tÃ©cnica simples Ã© usar todos os atributos e deixÃ¡-los fazer contribuiÃ§Ãµes em igual medida para as decisÃµes, admitindo que sejam  independentes entre si. 

Isso Ã© irreal, Ã© claro: o que torna os conjuntos de dados da vida real interessantes Ã© que os atributos certamente nÃ£o sÃ£o igualmente importantes ou independentes. Mas isso leva a um esquema simples que novamente funciona  surpreendentemente bem na prÃ¡tica.

Este mÃ©todo atende pelo nome de Naive Bayes, porque Ã© baseado na regra de Bayes e âingenuamenteâ assume independÃªncia â sÃ³ Ã© vÃ¡lido multiplicar probabilidades quando os eventos sÃ£o independentes. A suposiÃ§Ã£o de que os atributos sÃ£o independentes (dada a classe) na vida real certamente Ã© simplista. Mas, apesar do nome depreciativo,
Naive Bayes funciona muito bem quando testado em conjuntos de dados reais, especialmente quando combinados com alguns dos procedimentos de seleÃ§Ã£o de atributos que eliminam atributos redundantes e, portanto, nÃ£o independentes.

Dessa forma para uma nova instÃ¢ncia $x_k$ para a qual temos que determinar uma classe $y_k = \{c_1, c_2, \cdots, c_t\}$, vamos calcular a probabilidade da classe para cada um dos possÃ­veis valores $P(y_k = c_i | x_k)$. Desmembrando a instÃ¢ncia $x_k$ temos um conjunto de valoraÃ§Ãµes para cada um dos atributos: $x_k = (x_{k1}, x_{k2}, \cdots, x_{kn})$. 

Aplicando o Teorema de Bayes temos entÃ£o:

$P(y_k = c_i | \cap_{j=1}^{n} a_j = x_{kj}) = \frac{P(\cap_{j=1}^{n} a_j = x_{kj}|y_k = c_i) \cdot P(y_k = c_i)}{P((\cap_{j=1}^{n} a_j = x_{kj})}$

Assumindo a independÃªncia entre os atributos, dada a classe, temos que o termo 

$P(\cap_{j=1}^{n} a_j = x_{kj}|y_k = c_i) = \Pi_{j=1}^{n} P(a_j = x_{kj}|y_k = c_i)$

De acordo com a tabela de entrada, vamos nos concentrar portanto apenas nos exemplos onde a classe $c_i$ aparece e contabilizar a proporÃ§Ã£o em que cada atributo $a_j = x_{kj}$ para obtermos $P(a_j = x_{kj}|y_k = c_i)$.

Portanto conseguimos atravÃ©s desse modelo probabilÃ­stico simplificado, calcular qual a probabilidade (ou na prÃ¡tica, a *verossimilhanÃ§a*) de que um determinado exemplo pertenÃ§a a cada uma das classes. Evidentemente que o classificador Naive Bayes recomenda enfim a classe cuja probabilidade de pertinÃªncia seja a mais alta.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB

# Generate some random data
np.random.seed(0)
X1 = np.random.normal(loc=-2.0, scale=1.0, size=(100, 2))
X2 = np.random.normal(loc=2.0, scale=1.0, size=(100, 2))
X = np.vstack([X1, X2])
y = np.array([0] * 100 + [1] * 100)

# Fit the Naive Bayes classifier to the data
clf = GaussianNB()
clf.fit(X, y)

# Create a meshgrid to plot the decision boundary
h = 0.02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Plot the decision boundary
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
plt.title("Naive Bayes classifier on a 2D space")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

### Ãrvores ID3

O problema de construir uma Ã¡rvore de decisÃ£o pode ser expresso recursivamente. Primeiro, selecione um atributo para colocar no nÃ³ raiz e faÃ§a uma ramificaÃ§Ã£o para cada valor possÃ­vel. Isso divide o exemplo definido em subconjuntos, um para cada valor do atributo. Agora o processo pode ser repetido recursivamente para cada ramificaÃ§Ã£o, usando apenas as instÃ¢ncias que realmente alcanÃ§am a ramificaÃ§Ã£o. Se a qualquer momento todas as instÃ¢ncias de um nÃ³ tiverem a mesma classificaÃ§Ã£o, pare de desenvolver essa parte da Ã¡rvore.

A Ãºnica coisa que resta Ã© como determinar em qual atributo dividir, dado um conjunto de exemplos com diferentes classes.  Como nÃ³s procuramos pequenas Ã¡rvores, gostarÃ­amos que o atributo escolhido pudesse organizar ao mÃ¡ximo os ramos que ele define. Se tivÃ©ssemos uma medida da pureza de cada nÃ³, poderÃ­amos escolher o atributo que produz os nÃ³s filhos mais puros.

A medida de pureza que usaremos Ã© chamada de **entropia** e Ã© medida em unidades chamadas bits. Associado a um nÃ³ da Ã¡rvore, representa a quantidade esperada de informaÃ§Ãµes que seriam necessÃ¡rias para especificar se uma nova instÃ¢ncia deveria ser classificada como sim ou nÃ£o (supondo uma classificaÃ§Ã£o binÃ¡ria), dado que o exemplo atingiu aquele nÃ³. Ao contrÃ¡rio dos bits na memÃ³ria do computador, a quantidade esperada de informaÃ§Ãµes geralmente envolve fraÃ§Ãµes de um bit - e geralmente Ã© menor que um! 

Para ter uma ideia de quanta informaÃ§Ã£o precisamos para representar um conjunto de dados, podemos pensar no quÃ£o surpreendente Ã© encontrarmos um dado de cada tipo possÃ­vel. Ou seja, caso um valor seja muito comum, ele nÃ£o Ã© nada surpreendente, enquanto valores raros podem causar grande surpresa. A **surpresa** pode ser vista como algo inversamente proporcional Ã  probabilidade daquele valor. De fato vamos definir que a surpresa, por exemplo, da classe $y$ assumir um determinado valor $c_i$, Ã©: 

$surpresa(y=c_i) = log_2\ \frac{1}{P(y=c_i)}$. 

Para facilitar a representaÃ§Ã£o vamos usar $lg$ como $log_2$.

Finalmente vamos definir a **entropia portanto como o valor esperado da surpresa** do conjunto. Assim:

$Entropia(Y) = \sum_{i} P(y=c_i) \cdot surpresa(y=c_i)$

$Entropia(Y) = \sum_{i} P(y=c_i) \cdot lg\ \frac{1}{P(y=c_i)}$

 Pela definiÃ§Ã£o de logaritmo tambÃ©m podemos escrever $lg\ \frac{1}{P(y=c_i)} = -lg\ P(y=c_i)$, que Ã© a forma mais comum de encontrarmos a definiÃ§Ã£o de entropia:

$Entropia(Y) = - \sum_{i} P(y=c_i) \cdot lg\ P(y=c_i)$

Para medir portanto a contribuiÃ§Ã£o que um determinado atributo pode dar Ã  organizaÃ§Ã£o dos dados vamos definir a entropia relativa do atributo $X_j$ para a classe $Y$ como a entropia restante da classe para cada valor que o atributo $X_j$ possa assumir, ponderado pela proporÃ§Ã£o em que ele aparece.

$Entropia(X_j, Y) = \sum_{k \in D_j} P(X_j = k) \cdot Entropia(Y|X_j = k)$

A diferenÃ§a entre a entropia original e a entropia restante pela enumeraÃ§Ã£o de valores do atributo $X_j$ Ã© o chamado **ganho de informaÃ§Ã£o**:

$Ganho(X_j, Y) = Entropia(Y) - Entropia(X_j, Y)$

Finalmente essa Ã© a nossa medida de organizaÃ§Ã£o dos atributos. Em cada momento em que nosso algoritmo precisa rescursivamente definir qual Ã© o atributo que deve ser escolhido vamos calcular o ganho de informaÃ§Ã£o de cada um deles e escolher qual o atributo com a melhor estimativa organizativa dos dados.

```python
# ID3 (Iterative Dichotomiser 3) Algorithm implementation from scratch

import math
from collections import deque

class Node:
    """Contains the information of the node and another nodes of the Decision Tree."""

    def __init__(self):
        self.value = None
        self.next = None
        self.childs = None


class DecisionTreeClassifier:
    """Decision Tree Classifier using ID3 algorithm."""

    def __init__(self, X, feature_names, labels):
        self.X = X
        self.feature_names = feature_names
        self.labels = labels
        self.labelCategories = list(set(labels))
        self.labelCategoriesCount = [list(labels).count(x) for x in self.labelCategories]
        self.node = None
        self.entropy = self._get_entropy([x for x in range(len(self.labels))])  # calculates the initial entropy

    def _get_entropy(self, x_ids):
        """ Calculates the entropy.
        Parameters
        __________
        :param x_ids: list, List containing the instances ID's
        __________
        :return: entropy: float, Entropy.
        """
        # sorted labels by instance id
        labels = [self.labels[i] for i in x_ids]
        # count number of instances of each category
        label_count = [labels.count(x) for x in self.labelCategories]
        # calculate the entropy for each category and sum them
        entropy = sum([-count / len(x_ids) * math.log(count / len(x_ids), 2) if count else 0 for count in label_count])
        return entropy

    def _get_information_gain(self, x_ids, feature_id):
        """Calculates the information gain for a given feature based on its entropy and the total entropy of the system.
        Parameters
        __________
        :param x_ids: list, List containing the instances ID's
        :param feature_id: int, feature ID
        __________
        :return: info_gain: float, the information gain for a given feature.
        """
        # calculate total entropy
        info_gain = self._get_entropy(x_ids)
        # store in a list all the values of the chosen feature
        x_features = [self.X[x][feature_id] for x in x_ids]
        # get unique values
        feature_vals = list(set(x_features))
        # get frequency of each value
        feature_vals_count = [x_features.count(x) for x in feature_vals]
        # get the feature values ids
        feature_vals_id = [
            [x_ids[i]
            for i, x in enumerate(x_features)
            if x == y]
            for y in feature_vals
        ]

        # compute the information gain with the chosen feature
        info_gain = info_gain - sum([val_counts / len(x_ids) * self._get_entropy(val_ids)
                                     for val_counts, val_ids in zip(feature_vals_count, feature_vals_id)])

        return info_gain

    def _get_feature_max_information_gain(self, x_ids, feature_ids):
        """Finds the attribute/feature that maximizes the information gain.
        Parameters
        __________
        :param x_ids: list, List containing the samples ID's
        :param feature_ids: list, List containing the feature ID's
        __________
        :returns: string and int, feature and feature id of the feature that maximizes the information gain
        """
        # get the entropy for each feature
        features_entropy = [self._get_information_gain(x_ids, feature_id) for feature_id in feature_ids]
        # find the feature that maximises the information gain
        max_id = feature_ids[features_entropy.index(max(features_entropy))]

        return self.feature_names[max_id], max_id

    def id3(self):
        """Initializes ID3 algorithm to build a Decision Tree Classifier.
        :return: None
        """
        x_ids = [x for x in range(len(self.X))]
        feature_ids = [x for x in range(len(self.feature_names))]
        self.node = self._id3_recv(x_ids, feature_ids, self.node)
        print('')

    def _id3_recv(self, x_ids, feature_ids, node):
        """ID3 algorithm. It is called recursively until some criteria is met.
        Parameters
        __________
        :param x_ids: list, list containing the samples ID's
        :param feature_ids: list, List containing the feature ID's
        :param node: object, An instance of the class Nodes
        __________
        :returns: An instance of the class Node containing all the information of the nodes in the Decision Tree
        """
        if not node:
            node = Node()  # initialize nodes
        # sorted labels by instance id
        labels_in_features = [self.labels[x] for x in x_ids]
        # if all the example have the same class (pure node), return node
        if len(set(labels_in_features)) == 1:
            node.value = self.labels[x_ids[0]]
            return node
        # if there are not more feature to compute, return node with the most probable class
        if len(feature_ids) == 0:
            node.value = max(set(labels_in_features), key=labels_in_features.count)  # compute mode
            return node
        # else...
        # choose the feature that maximizes the information gain
        best_feature_name, best_feature_id = self._get_feature_max_information_gain(x_ids, feature_ids)
        node.value = best_feature_name
        node.childs = []
        # value of the chosen feature for each instance
        feature_values = list(set([self.X[x][best_feature_id] for x in x_ids]))
        # loop through all the values
        for value in feature_values:
            child = Node()
            child.value = value  # add a branch from the node to each feature value in our feature
            node.childs.append(child)  # append new child node to current node
            child_x_ids = [x for x in x_ids if self.X[x][best_feature_id] == value]
            if not child_x_ids:
                child.next = max(set(labels_in_features), key=labels_in_features.count)
                print('')
            else:
                if feature_ids and best_feature_id in feature_ids:
                    to_remove = feature_ids.index(best_feature_id)
                    feature_ids.pop(to_remove)
                # recursively call the algorithm
                child.next = self._id3_recv(child_x_ids, feature_ids, child.next)
        return node

    def printTree(self):
        if not self.node:
            return
        nodes = deque()
        nodes.append(self.node)
        while len(nodes) > 0:
            node = nodes.popleft()
            print(node.value)
            if node.childs:
                for child in node.childs:
                    print('({})'.format(child.value))
                    nodes.append(child.next)
            elif node.next:
                print(node.next)
```

#### RegressÃ£o linear

Quando o resultado, ou classe, Ã© numÃ©rico e todos os atributos sÃ£o numÃ©ricos, a regressÃ£o linear Ã© uma tÃ©cnica natural a ser considerada. Este Ã© um mÃ©todo bÃ¡sico em estatÃ­stica. A ideia Ã© expressar a classe como uma combinaÃ§Ã£o linear dos atributos, com pesos prÃ©-determinados:

$y = w_0 + w_1 x_1 + w_2 x_2 + \cdots + w_n x_n$

Os pesos sÃ£o calculados a partir dos dados de treinamento. Aqui a notaÃ§Ã£o fica um pouco pesada, pois precisamos de uma forma de expressar os valores dos atributos para cada instÃ¢ncia de treinamento. Vamos utilizar $x_k^{(i)}$ como o valor do $k$-Ã©simo atributo da instÃ¢ncia $i$. AlÃ©m disso, Ã© notacionalmente conveniente assumir um atributo extra $a_0$, cujo valor Ã© sempre 1.

Portanto o valor previsto da classe pela instÃ¢ncia $i$ Ã©:

$y = w_0a_0^{(i)} + w_1 x_1^{(i)} + w_2 x_2^{(i)} + \cdots + w_n x_n^{(i)} = \sum \limits_{k=0}^{n} w_k x_k^{(i)}$

Contudo a prÃ³pria instÃ¢ncia $i$ traz um valor para a classe, $y^{(i)}$. Pode-se perceber a qualidade dos pesos calculados, portanto pela diferenÃ§a entre a classe de fato e a classe prevista. 

O mÃ©todo de regressÃ£o linear de mÃ­nimos quadrados consiste em escolher os coeficientes $w_k$ - existem $n+1$ deles - para minimizar a soma dos quadrados dessas diferenÃ§as em todas as instÃ¢ncias de treinamento. Suponha que existam $m$ instÃ¢ncias de treinamento, entÃ£o a soma dos quadrados das diferenÃ§as Ã©:

$\sum \limits_{i=1}^{m} \left( y^{(i)} - \sum \limits_{k=0}^{n} w_k x_k^{(i)} \right)^2$

onde a expressÃ£o dentro dos parÃªnteses Ã© a diferenÃ§a entre a classe real da $i$-Ã©sima instÃ¢ncia e sua classe prevista. Essa soma de quadrados Ã© o que temos que minimizar escolhendo os coeficientes apropriadamente.

A regressÃ£o linear Ã© um mÃ©todo excelente e simples de previsÃ£o numÃ©rica e tem sido amplamente utilizada em aplicaÃ§Ãµes estatÃ­sticas hÃ¡ dÃ©cadas. Claro, modelos lineares bÃ¡sicos sofrem da desvantagem de, bem, linearidade. Se os dados exibirem uma dependÃªncia nÃ£o linear, a linha reta de melhor ajuste serÃ¡ encontrada, onde âmelhorâ Ã© interpretado como a menor diferenÃ§a mÃ©dia quadrÃ¡tica. Esta linha pode nÃ£o se encaixar muito bem. No entanto, os modelos lineares servem bem como blocos de construÃ§Ã£o ou pontos de partida para mÃ©todos de aprendizado mais complexos.






```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(0)
X = np.random.uniform(low=0, high=10, size=50)
y = 2*X + 1 + np.random.normal(loc=0.0, scale=2.0, size=50)

# Fit a linear regression model
model = np.polyfit(X, y, 1)
predict = np.poly1d(model)

# Compute the predicted values and errors
y_pred = predict(X)
errors = abs(y - y_pred)/2

# Plot the data points and regression line with error bars
fig, ax = plt.subplots()
ax.scatter(X, y, color='blue')
ax.plot(X, predict(X), color='red')
ax.errorbar(X, (y+y_pred)/2, yerr=errors, fmt='none', color='black', alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('y')
ax.set_title('Regression plot with error bars')

plt.show()
```

#### RegressÃ£o logÃ­stica

A regressÃ£o linear pode ser facilmente usada para classificaÃ§Ã£o em domÃ­nios com atributos numÃ©ricos. De fato, podemos usar qualquer tÃ©cnica de regressÃ£o para classificaÃ§Ã£o. O truque Ã© realizar uma regressÃ£o para cada classe, definindo a saÃ­da igual a um para instÃ¢ncias de treinamento que pertencem Ã  classe e zero para aquelas que nÃ£o pertencem. O resultado Ã© uma expressÃ£o linear para a classe. Em seguida, dado um exemplo de teste de classe desconhecida, calcule o valor de cada expressÃ£o linear e escolha a que for maior. Quando usado com regressÃ£o linear, esse esquema Ã s vezes Ã© chamado de regressÃ£o linear multiresposta.

Uma maneira de olhar para a regressÃ£o linear multiresposta Ã© imaginar que ela se aproxima de uma funÃ§Ã£o de pertinÃªncia numÃ©rica para cada classe. A funÃ§Ã£o de pertinÃªncia Ã© 1 para instÃ¢ncias que pertencem a essa classe e 0 para outras instÃ¢ncias. Dada uma nova instÃ¢ncia, calculamos seus membros para cada classe e selecionamos a maior.

A regressÃ£o linear multiresposta geralmente produz bons resultados na prÃ¡tica. No entanto, tem duas desvantagens. Primeiro, os valores de associaÃ§Ã£o que ela produz nÃ£o sÃ£o probabilidades adequadas porque podem ficar fora do intervalo $[0,1]$. Em segundo lugar, a regressÃ£o dos mÃ­nimos quadrados assume que os erros nÃ£o sÃ£o apenas estatisticamente independentes, mas tambÃ©m sÃ£o normalmente distribuÃ­dos com o mesmo desvio padrÃ£o, uma suposiÃ§Ã£o que Ã© flagrantemente violada quando o mÃ©todo Ã© aplicado a problemas de classificaÃ§Ã£o porque as observaÃ§Ãµes sempre assumem apenas os valores 0 e 1.

Uma tÃ©cnica estatÃ­stica relacionada chamada regressÃ£o logÃ­stica nÃ£o sofre desses problemas. Em vez de aproximar os valores 0 e 1 diretamente, arriscando assim valores de probabilidade ilegÃ­timos quando o alvo Ã© ultrapassado, a regressÃ£o logÃ­stica constrÃ³i um modelo linear baseado em uma variÃ¡vel alvo transformada.

Suponha primeiro que existam apenas duas classes. A regressÃ£o logÃ­stica substitui a variÃ¡vel de destino original

$P(1|x_1, x_2, \cdots, x_n)$

que nÃ£o pode ser aproximada com precisÃ£o usando uma funÃ§Ã£o linear, por

$\log\ \frac{P(1|x_1, x_2, \cdots, x_n)}{1 - P(1|x_1, x_2, \cdots, x_n)}$

A variÃ¡vel transformada Ã© aproximada usando uma funÃ§Ã£o linear como as geradas por regressÃ£o linear. O modelo resultante Ã©:

$P(1|x_1, x_2, \cdots, x_n) = \frac{1}{1 + \exp(-(w_0 + w_1 x_1 + w_2 x_2 + \cdots + w_n x_n))}$

Assim como na regressÃ£o linear, devem ser encontrados pesos que se ajustem bem aos dados de treinamento. A regressÃ£o linear mede a qualidade do ajuste usando o erro quadrado. Na regressÃ£o logÃ­stica, a verossimilhanÃ§a logarÃ­tmica do modelo Ã© usada. Isso Ã© dado por:

$\sum \limits_{i=1}^{m} (1 - y^{(i)})\log(1-P(1|x_1^{(i)}, x_2^{(i)}, \cdots, x_n^{(i)}) + y^{(i)}\log P(1|x_1^{(i)}, x_2^{(i)}, \cdots, x_n^{(i)})$

Como $y^{(i)} \in \{0,1\}$ a funÃ§Ã£o acima pode ser resolvida da mesma forma dos mÃ­nimos quadrados para regressÃ£o linear.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Generate a random dataset with two classes
np.random.seed(0)
mean1 = [45, 45]
cov1 = [[30,0],[0,30]]
mean2 = [65, 65]
cov2 = [[40, 20], [20, 40]]

X1 = np.random.multivariate_normal(mean1, cov1, 50)
X2 = np.random.multivariate_normal(mean2, cov2, 50)
X = np.concatenate((X1, X2))
y = np.concatenate([np.zeros(50), np.ones(50)])

# Create the logistic regression classifier
clf = LogisticRegression()

# Fit the classifier to the data
clf.fit(X, y)

# Plot the decision boundary
x1_min, x1_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
x2_min, x2_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max, 100), np.linspace(x2_min, x2_max, 100))
Z = clf.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z = Z.reshape(xx1.shape)
plt.contourf(xx1, xx2, Z, cmap=plt.cm.RdBu_r, alpha=0.6)

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu_r)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear Classifier')
plt.show()
```
